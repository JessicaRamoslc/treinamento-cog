# Tratando dados:

função recebe um dicionário

```python
filmes = filmes.fillna({'investimento': filmes['investimento'].mean(), 'bilheteria': filmes['bilheteria'].median()}) #conseguimos a media
```

```python
filmes
```

```python
filmes = filmes.dropna() #dropna retira a linha que tiver algum valor em branco no arquivo
```

```python
filmes
```

```python
filmes = pd.get_dummies(filmes, columns=['genero']) #atribuir um número (0,1) a variável #tratando variavel categorica
```

# Dividir entre treino e testes:

x é o que vai aprender, y é o que ele vai prever

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = filmes['investimento']
```

```python
y = filmes['bilheteria']
```

```python
x
```

```python
y
```

x treino, x de teste, y de treino e y de teste

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)#25% para teste, 75% para treino #treino e teste
```

# Treinamento

regressão linear

```python
import numpy as np
```

```python
x_train = np.array(x_train).reshape(len(x_train), 1)
x_test = np.array(x_test).reshape(len(x_test), 1)
y_test = np.array(y_test).reshape(len(y_test), 1)
y_train = np.array(y_train).reshape(len(y_train), 1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
modelo = LinearRegression() #instanciar a classe
```

```python
modelo.fit(x_train, y_train)
```

```python
modelo.score(x_test, y_test)
```
